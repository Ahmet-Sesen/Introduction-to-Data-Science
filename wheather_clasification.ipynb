```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('weather_classification_data.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
unique_val_wheather_type=df['Weather Type'].unique()
print(unique_val_wheather_type)
```

```python
weather_counts=df['Weather Type'].value_counts()
print(weather_counts)
```

```python
weather_counts.plot(kind='bar')

plt.title('Distribution of Weather Types')
plt.xlabel('Weather Type')
plt.ylabel('Count')

plt.show()
```

```python
df_above_30=df[df['Temperature']>30]

df_above_30.head()
```

```python
average_temperature_by_seasen=df.groupby('Season')['Temperature'].mean()
print(average_temperature_by_seasen)
```

```python
plt.scatter(df['Temperature'], df['Humidity'])

plt.title('Scatter Plot of Temperature vs Humuity')
plt.xlabel('Temperature (%C)')
plt.ylabel('Humidity (%)')

plt.show()
```

```python
df['Feels Like'] = df['Temperature'] - df['Wind Speed'] / 10
df.head()
```

```python
df_sorted=df.sort_values(by='Temperature',ascending=False)

df_sorted.head()
```

```python
df['Temperature (F)']=df['Temperature']*(9/5) +32

df.head()
```

```python
filtered_wind_spead=df[(df['Wind Speed'] >= 10) & (df['Wind Speed'] <=20)]
```

```python
average_visibility=df.groupby('Cloud Cover')['Visibility (km)'].mean()
print(average_visibility)
```

```python
df['Precipitation (%)']=df['Precipitation (%)'].where(df['Precipitation (%)'] >=0,0)
```

```python
df['UV Index High']=df['UV Index'] > 6
df.head()
```

```python
df.rename(columns={'Atmospheric Pressure': 'Pressure (hPa)'},inplace=True)
df.head()
```

```python
record_counts_season=df['Season'].value_counts()
print(record_counts_season)
```

```python
duplicates=df.duplicated().sum()
print(f"Number of duplicates:{duplicates}")
```

```python
pivot_table=df.pivot_table(values='Temperature', index='Season',columns='Location',aggfunc='mean')
print(pivot_table)
```

```python
weather_counts=df['Weather Type'].value_counts()
print(weather_counts)
```

```python
max_wind_speed_by_weather_type=df.groupby('Weather Type')['Wind Speed'].max()

print(max_wind_speed_by_weather_type)
```

```python
sum_precipitation_by_location= df.groupby('Location')['Precipitation (%)'].sum()

print(sum_precipitation_by_location)
```

```python
plt.hist(df['Temperature'], bins=10, edgecolor='black')

plt.title('Histogram of Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')

plt.show()
```

```python
max_humidity_index=df['Humidity'].idxmax()

df.loc[max_humidity_index]
```

```python
df_sorted_visibility=df.sort_values(by='Visibility (km)',ascending=False)
top_10_visibility=df_sorted_visibility.head(10)

top_10_visibility
```

```python
temperature_humidity_df = df[['Temperature', 'Humidity']]

print(temperature_humidity_df.head())
```

```python
numeric_df = df.select_dtypes(include='number')

categorical_df = df.select_dtypes(include='object')

print("Numeric DataFrame:")
print(numeric_df.head())

print("\nCategorical DataFrame:")
print(categorical_df.head())
```
