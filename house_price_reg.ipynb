## Outlier Detection

```python
def replace_outliers_iqr(df):
    df_replaced = data.copy()  
    outliers_count = {}  # Dictionary to store count of outliers for each column
    for col in df_replaced.select_dtypes(include=np.number).columns:  # Apply to numeric columns only
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df_replaced[col].quantile(0.25)
        Q3 = df_replaced[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # Calculate lower and upper boundaries
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Count the outliers
        outliers_below = df_replaced[col] < lower_bound
        outliers_above = df_replaced[col] > upper_bound
        total_outliers = outliers_below.sum() + outliers_above.sum()
        
        # Store count of outliers for each column
        outliers_count[col] = total_outliers
        
        # Replace outliers with nearest boundary
        df_replaced[col] = np.where(df_replaced[col] < lower_bound, lower_bound,
                                    np.where(df_replaced[col] > upper_bound, upper_bound, df_replaced[col]))
        
        for col, count in outliers_count.items():
            print(f"Column '{col}': {count} outliers detected.")

    return df_replaced
```

```python
# Apply the function to replace outliers
data = replace_outliers_iqr(data)
```

```python
def combine_rare_categories(df, column, threshold=80):
    # Get category counts
    category_counts = df[column].value_counts()
    
    # Define rare categories
    rare_categories = category_counts[category_counts < threshold].index
    
    # Replace rare categories with 'Other'
    df[column] = df[column].apply(lambda x: 'Other' if x in rare_categories else x)
    return df

# Apply to columns
columns_to_adjust = data.select_dtypes(include=['object']).columns.tolist()
columns_to_adjust.remove("Neighborhood")

for col in columns_to_adjust:
    data = combine_rare_categories(data, col)
```

## Linear Regression
